```python
#we can not to do like arr1 * arr2 like diff dimensions but we can use the
#matmul for the multiplication of two diff dimension arrays
a = np.ones((2,3))
b = np.full((3,2), 5)

new = np.matmul(a,b)
print(new)
```

Reorganizing an array

```python
#reshaping the arrays while reshaping the new shape should fit the all the elements of previous shape data no less and no morev1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])
before = np.array([[2,3,4,4],[2,4,5,6]])

after = before.reshape((2,4))
after
```

```python
# Vertically stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2,v1,v2])
```

```python
#horizontal stacking of vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.hstack([v1,v2,v1,v2])
```
